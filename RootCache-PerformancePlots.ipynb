```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

width = .35 # width of a bar

# Speculated max capacity of tree: 2441406 nodes (Given degree: 3, height: 12)
# 
# 1M Searches - Pure Randomised 
# Shows significant improvement until cache size of 2K
# Baseline Iterated Levels: 12000000 (12 Levels)
# no of levels saved - histogram (Queries) TODO
# Max Iterations Saved (2K Cache Size) - 26.715 % levels Saved
# We don't see a considerable performance improvement in the iterated levels saved from 1 to 64 cache size.
# reasons for mediocre performance improvement: 1. Duplication of the entries in root cache 2. due to constant switching of keys being searched in the B+ Tree
m1_t = pd.DataFrame({
 'baseline' : [1, 1, 1, 1, 1, 1, 1],
 'iterated_levels' : [0.999986167,0.9630743,0.93407134,0.893579834,0.84463075,0.790406,0.73284925],
 'hit_rate' : [0.0056,61.9207,77.9543,88.0176,93.8519,97.0544,98.628100]})

m1_t[['iterated_levels']].plot(kind='bar', color='green', width = width)
m1_t['baseline'].plot(kind='line', marker='*', color='black', ms=10)

ax = plt.gca()
plt.xlim([-width, len(m1_t['hit_rate'])-width])
ax.set_xticklabels(('1', '64', '128', '256', '512', '1024', '2048'))

plt.show()
```

#### The number of levels iterated decrease with the increase in cache size and irregularity in searches.
